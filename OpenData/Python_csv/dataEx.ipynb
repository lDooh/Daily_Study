+ `matplotlib` 모듈은 파이썬에서 데이터를 시각화 할 때 사용하는 라이브러리이다.

```python
import matplotlib.pyplot as plt
```

+ `plot()` 함수를 이용해 데이터를 입력
    + `label` 속성으로 범례 표기 가능, `legend()` 함수로 출력
    + `linestyle` 또는 `ls`속성으로 선 모양 변경 가능, 기본값은 \`-\`
+ `xlabel()`, `ylabel()` 함수로 레이블 표시 가능
    + `labelpad` 속성은 표와 레이블 사이의 여백(padding)을 지정, 기본값은 4
    + `loc` 속성은 레이블의 위치 설정, x축은 left, center, right, y축은 bottom, center, top
    + `fontdict` 속성으로 폰트 지정 가능
+ `legend()` 함수로 범례 표시
    + `loc` 속성으로 표시 위치 지정 가능
    + `ncol` 속성으로 표시될 열의 개수 지정 가능
    + `frameon` 속성으로 상자의 테두리를 표시할지 말지 지정, 기본값은 True
    + `shadow` 속성으로 상자의 그림자 지정 가능
+ `axis()` 함수로 x축과 y축의 최소-최대값 지정 [xmin, xmax, ymin, ymax]
    + `xlim()` 함수와 `ylim()` 함수를 이용해 따로 지정할 수 있음
+ `show()` 함수를 이용해 그래프를 화면에 출력

```python
plt.title("Consumer Price Index")
plt.plot(cpi_dataframe['2021. 01'], label = "Jan.", color="r", ls="--")
plt.plot(cpi_dataframe['2021. 02'], label = "Feb.", ls=":")
plt.plot(cpi_dataframe['2021. 03'], label = "Mar.", ls="-.")
plt.xlabel("x-axis", labelpad = 30, loc="right", fontdict={"family": "serif", "color": "r", "weight": "bold", "size": 20})
plt.ylabel("y-axis", labelpad = 10)
plt.legend(loc="upper right", ncol=2, fontsize=12, frameon=True, shadow=True)
plt.axis([1, 15, 100, 103])
plt.show()
```

+ 범례의 위치는 `legend()` 함수의 `loc` 속성으로 지정할 수 있다.
    + 0~10까지의 정수를 속성의 값으로 넣어서 위치를 지정할 수도 있다.
